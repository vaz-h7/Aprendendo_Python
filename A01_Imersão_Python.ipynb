```python
df.columns = [
    'ano',
    'seneoridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'usd',
    'residencia',
    'remoto',
    'empresa',
    'tamanho_empresa'
]

print(df.columns)
```

```python
df['seneoridade'] = df['seneoridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
df['seneoridade'].value_counts()
```

```python
#Acima, utilizamos IA novamente para traduzir para português as siglas das colunas da tabela
#Abaixo, repeti o mesmo processo para as demais
#SEMPRE LEMBRAR DE VERIFICAR SE ESTÁ TUDO CORRETO APÓS USO DA IA!!!!!
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
print("Contrato (atualizado):")
print(df['contrato'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})
print("\nRemoto (atualizado):")
print(df['remoto'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})
print("\nTamanho da Empresa (atualizado):")
print(df['tamanho_empresa'].value_counts())
```

```python
df['cargo'] = df['cargo'].replace({
    'Data Engineer': 'Engenheiro de Dados',
    'Data Scientist': 'Cientista de Dados',
    'Data Analyst': 'Analista de Dados',
    'Machine Learning Engineer': 'Engenheiro de Machine Learning',
    'Analytics Engineer': 'Engenheiro de Análise',
    'Research Scientist': 'Cientista de Pesquisa',
    'Applied Scientist': 'Cientista Aplicado',
    'ML Engineer': 'Engenheiro de ML',
    'Cloud Engineer': 'Engenheiro de Nuvem',
    'AI Engineer': 'Engenheiro de IA',
    'Software Engineer': 'Engenheiro de Software',
    'Data Architect': 'Arquiteto de Dados',
    'Solutions Engineer': 'Engenheiro de Soluções'
})
print(df['cargo'].value_counts().head(10))
```
