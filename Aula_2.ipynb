```python
import numpy as np #Importando a biblioteca numpy e apelidando para np.
df_salarios = pd.DataFrame({
    'nome': ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    'salário': [4000, np.nan, 5000, np.nan, 100000]
}) #Criando um dataframe com os dados acima, np.nan significa valor nulo.

df_salarios['salario_media'] = df_salarios['salário'].fillna(df_salarios['salário'].mean().round(2))
#O comando acima serve para criar uma nova coluna, 'salario_media'. Fillna para preencher valores nulos com a média(mean) dos salários e round(2) para arredondar com duas casas decimais.
df_salarios['salario_mediana'] = df_salarios['salário'].fillna(df_salarios['salário'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    #Criando outro dataframe para exemplificar comandos.
    "Dias da semana" : ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura' : [30, np.nan, np.nan, 28, 27]
})
#ffill preenche com o valor anterior ao nulo, neste caso, 30.
df_temperaturas['Preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
#bfill preenche com o valor posterior ao nulo.
df_temperaturas['Preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```
