```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='seneoridade', y='usd')
# O primeiro comando é da própria biblioteca, para gerar um gráfico de colunas.
# Data = 'base de dados' e após isso selecionar as colunas que irão em cada eixo.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5)) # Definição do tamanho da figura.
sns.barplot(data=df_limpo, x='seneoridade', y='usd') # Gerar gráfico.
# Definição do título e rótulos de eixos.
plt.title('Salário médio por nível de seneoridade')
plt.xlabel('Seneoridade')
plt.ylabel('Salário médio anual em USD')
plt.show() # Para apresentar a imagem
```

```python
# Usamos o comando groupby na base de dados para agrupar e ordenar a média das colunas
df_limpo.groupby('seneoridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('seneoridade')['usd'].mean().sort_values(ascending=True).index
#Para salvar em uma variável, True = menor ao maior, False = maior ao menor.
ordem
```

```python
plt.figure(figsize=(8, 5)) # Definição do tamanho da figura.
sns.barplot(data=df_limpo, x='seneoridade', y='usd', order=ordem) # Gerar gráfico.
# Definição do título e rótulos de eixos.
plt.title('Salário médio por nível de seneoridade')
plt.xlabel('Seneoridade')
plt.ylabel('Salário médio anual em USD')
plt.show() # Para apresentar a imagem
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df_limpo['usd'], bins = 50, kde = True)
plt.title('Distribuição dos salários anuais')
plt.xlabel('Salário em USD')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_limpo['usd'])
plt.title('Boxplot Salário')
plt.xlabel('Salário em USD')
plt.show()
```

```python
ordem_seneoridade = ['Júnior', 'Pleno', 'Sênior', 'Executivo']
plt.figure(figsize=(8, 5))
sns.boxplot(x='seneoridade', y='usd', data=df_limpo, order=ordem_seneoridade)
plt.title('Distribuição salarial por seneoridade')
plt.xlabel('Salário em USD')
plt.ylabel('USD')
plt.show()
```

```python
ordem_seneoridade = ['Júnior', 'Pleno', 'Sênior', 'Executivo']
plt.figure(figsize=(8, 5))
sns.boxplot(x='seneoridade', y='usd', data=df_limpo, order=ordem_seneoridade, palette='Set2', hue='seneoridade')
# Foi adicionado os parâmetros 'palette' para paleta de cor e 'hue' para especificar onde irão as cores.
plt.title('Distribuição salarial por seneoridade')
plt.xlabel('Salário em USD')
plt.ylabel('USD')
plt.show()
```

```python
import plotly.express as px
```

```python
df_media_salario_seneoridade = df_limpo.groupby('seneoridade')['usd'].mean().reset_index()

fig = px.bar(
    df_media_salario_seneoridade,
    x='seneoridade',
    y='usd',
    title='Salário Médio Anual por Nível de Senioridade',
    labels={'seneoridade': 'Nível de Senioridade', 'usd': 'Salário Médio Anual em USD'},
    color='seneoridade',
    category_orders={'seneoridade': ['Júnior', 'Pleno', 'Sênior', 'Executivo']}
)

fig.update_layout(xaxis_title='Senioridade', yaxis_title='Salário Médio Anual (USD)')
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']
fig = px.pie(
    remoto_contagem,
    names = 'tipo_trabalho',
    values = 'quantidade',
    title= 'Proporção dos Tipos de Trabalho'
)
fig.show()
```
